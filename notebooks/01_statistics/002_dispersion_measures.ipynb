# DISPERSION MEASURES - VARIATION | STANDARD DEVIATION | IQR

```python
import numpy as np

data = np.array([2.59, 2.64, 2.60, 2.62, 2.57, 2.55, 2.61, 2.50, 2.63, 2.64])

#mean
mean = np.mean(data).round(2)

variation = np.var(data).round(2)
standard = np.std(data).round(2)

q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)

iqr = (q3 - q1).round(2)

print(f'Mean: {mean}')
print(f'Variation: {variation}')
print(f'Standard Deviation: {standard}')
print(f'IQR: {iqr}')

```
