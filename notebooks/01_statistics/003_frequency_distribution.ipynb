#Frequency Distribution with NumPy and Pandas

```python
import pandas as pd
import numpy as np

data = np.array([1,2,2,3,3,3,4,4,5])
```

```python
#Identify the absolute frequency

freq = pd.Series(data).value_counts().sort_index()
print(freq)
```

```python
#Relative Frequency (%)

freq_rel = pd.Series(data).value_counts(normalize=True).sort_index()*100
print(freq_rel)
```

```python
import matplotlib.pyplot as plt

freq.plot(kind='bar')
plt.title('Frequency Distribution')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```
