**Exercise 1**: Calculation with Raw Data (Interest Rates)

The interest rates received by 10 stocks over a certain period were:
2.59; 2.64; 2.60; 2.62; 2.57; 2.55; 2.61; 2.50; 2.63; 2.64

- Calculate the arithmetic mean, median, and mode for this dataset.
- Remember that to compute the median, the values must be sorted first.


```python
def calculate_central_measures(*args):

    data = list(args)
    sorted_data = sorted(data)

    total = sum(sorted_data)
    mean = total / len(sorted_data)

    if len(sorted_data) % 2 == 0:
        n = len(sorted_data)

        middle1 = sorted_data[n // 2 - 1]
        middle2 = sorted_data[n // 2]

        median = (middle1 + middle2) / 2

    else:
        n = len(sorted_data)
        median = sorted_data[n // 2]

    return print(f'Mean: {mean} | Median: {median}')


calculate_central_measures(2.59, 2.64, 2.60, 2.62, 2.57, 2.55, 2.61, 2.50, 2.63, 2.64)
```
