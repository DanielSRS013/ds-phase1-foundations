##SECTION 1 - THE ndarray (n-dimensional array)

Things that define ndarray:

**Shape** - the dimensions of the array. A 1D array of 6 numbers has shape (6,). A 2D array with 3 rows and 4 columns has shape (3,4).
PS.: SHAPE IS ALWAYS A TUPLE

**Dtype** - the data type of every element.

**Data buffer** - the raw block of memory where the values live.

```python
# importing the numpy

import numpy as np

# Creating arrays = the most common ways

a = np.array([1,2,3,4,5])         # from a Python list
b = np.array([1.5, 2.5 , 3.5])    # float64 automatically
c = np.zeros((3,4))               # 3x4 array of zeros (3 rows and 4 columns)
d = np.ones((2,3))                # 2x3 array of ones
e = np.arange(0, 20 ,2)           # like range() but returns array
f = np.linspace(0,1,5)            # 5 evenly spaced values from 0 to 1
g = np.full((2,2), 7)             # 2x2 array filled with 7
h = np.eye(3)                     # 3x3 identity matrix

# Inspecting an array
print(a.shape)  # number of rows and columns (5,)
print(a.dtype)  # data type of the array (int)
print(a.ndim)   #dimensions number
print(a.size)   #total number of elements
```

## SECTION 2 - ARITHMETIC AND VECTORIZATION

Core idea of Numpy. No loop required to do any arithmetic operation between two arrays of the same shape.

```python
a = np.array([1,2,3,4])
b = np.array([10,20,30,40])

print(a+b)
print(a*b)
print(b/a)
print(a**2)
```

Scalar Operations broadcast automatically

```python
a = np.array([1,2,3,4])
print(a*2)
print(a+100)
print(a>2)
```

Universal Functions - Math Functions

```python
a = np.array([1.0, 4.0, 9.0, 16.0])

print(np.sqrt(a))
print(np.log(a))                     # natural log of each element
print(np.exp(a))                     # e^x for each element
print(np.abs(np.array([-3,-1,2,4]))) #absolute value
```
